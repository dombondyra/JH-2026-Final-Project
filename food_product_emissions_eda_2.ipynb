Goals of this notebook

```python
import os
import pandas as pd

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(os.path.join('data', 'Food_Product_Emissions.csv'))
```

## Add food category column ( plant based / animal based )
Food are sorted based on primarily animal or plant based 

```python
animal_based = ['Beef (beef herd)', 'Beef (dairy herd)', 'Cheese', 'Eggs', 'Fish (farmed)', 'Lamb & Mutton', 
                'Milk', 'Pig Meat', 'Poultry Meat', 'Shrimps (farmed)']

plant_based = ['Apples', 'Bananas', 'Barley', 'Beet Sugar', 'Berries & Grapes', 'Brassicas', 'Cane Sugar', 
               'Cassava', 'Citrus Fruit', 'Coffee', 'Dark Chocolate', 'Groundnuts', 'Maize', 'Nuts', 'Oatmeal', 
               'Olive Oil', 'Onions & Leeks', 'Other Fruit', 'Other Pulses', 'Other Vegetables', 'Palm Oil', 
               'Peas', 'Potatoes', 'Rapeseed Oil', 'Rice', 'Root Vegetables', 'Soybean Oil', 'Soymilk', 
               'Sunflower Oil', 'Tofu', 'Tomatoes', 'Wheat & Rye', 'Wine']

df.loc[df['Food product'].isin(animal_based), 'category'] = 'animal'
df.loc[df['Food product'].isin(plant_based), 'category'] = 'plant'
```

```python
counts = df['category'].value_counts(normalize=True)

plt.bar(counts.index, counts.values)
plt.title("Distribution of food type (animal vs protein)")
plt.show()
```
