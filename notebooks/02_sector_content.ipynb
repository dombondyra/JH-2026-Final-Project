```python
import os

import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px

from sklearn.linear_model import LinearRegression


data_dir = os.path.join(os.getcwd(), '..', 'data')
```

### Import and Clean Data

Import data, drop unnecessary columns, convert to a dataframe that is easier to plot:

```python
# clean data make sure there is no missing data, make sure the periods are steady
df = pd.read_csv(os.path.join(data_dir, 'ghg-emissions.csv'))
# sectors = df['Sector'].unique()
unit = df['unit'].unique()[0]

sectors = list(df['Sector'])
# sectors[0] = 'Year'
df = df.drop(columns=['unit', 'Sector'])
df = df.T
df.columns= sectors

df['Year'] = df.index.astype(int)
```

Ensure there is no missing values:

```python
are_any_na = df.isna().any(axis=None)
if are_any_na == False:
    print("There are no missing values in the data.")
else: 
    print("ERROR: The data contains null values.")

# ensure there are no missing years
year_start = df['Year'].iloc[0]
year_end = df['Year'].iloc[-1]
total_num_years = int(year_end) - int(year_start) + 1 # +1 is added becasue the years are inclusive
if total_num_years == len(df):
    print("There are no missing years in the data.")
else:
    print("ERROR: The data is missing data for at least one year")
```

### Plot: GHG Emissions from Agriculture 1990-2023

```python
fig = px.line(df, x='Year', y='Agriculture')
fig.update_layout(title="GHG Emissions from Agriculture",
                  yaxis_title=unit)
fig.update_xaxes(tickangle=45)
fig.show()
```

### Linear Regression to Predict Future Emissions 

```python
# set data
X = df[['Year']]
y = df['Agriculture']

# fit the model
model = LinearRegression()
model.fit(X, y)

# get model attributes
slope = model.coef_[0]
intercept = model.intercept_
r_squared = model.score(X, y)

print(f"Equation: y = {slope:.2f}x + {intercept:.2f}")
print(f"R² Score: {r_squared:.4f}")
```

```python
# predict data for the next 20 years (10, 20, 50?)
num_year_predict = 20
predict_start = year_end 
predict_years = list(range(predict_start, predict_start + num_year_predict + 1))

df_predict = pd.DataFrame({'Year': predict_years}, index = predict_years)


predictions = model.predict(df_predict[['Year']])
df_predict['Agriculture'] = predictions
```

```python
# plot data with predictions 

fig.add_scatter(
        x=df_predict['Year'], 
        y=df_predict['Agriculture'], 
        mode='lines', 
        line=dict(color='red'),
        name='Predicted Emissions'
)
fig.show()
```
