## Comparing Sectors

```python
import os

import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px

data_dir = os.path.join(os.getcwd(), 'data')
```

# GHG Emissions Data EDA

```python
df = pd.read_csv(os.path.join(data_dir, 'ghg-emissions.csv'))
# sectors = df['Sector'].unique()
unit = df['unit'].unique()[0]

sectors = list(df['Sector'])
# sectors[0] = 'Year'
df = df.drop(columns=['unit', 'Sector'])
df = df.T
df.columns= sectors

df['Year'] = df.index
```

```python
df
```

```python
fig = px.line(df, x='Year', y=sectors)
fig.update_layout(title="GHG Emissions per Sector",
                  yaxis_title=unit)
fig.update_xaxes(tickangle=45)
fig.show()
```

```python
YEAR = '2023'
df[df['Year'] == YEAR]['Agriculture'] 
# df
em_per_sector = {}
for sector in sectors:
    em_per_sector[sector] = round(float(df[df['Year'] == YEAR][sector].values[0]), 2)

# # simplify name
em_per_sector['Land-Use/Forestry'] = em_per_sector.pop('Land Use, Land-Use Change and Forestry')

# sort 
em_per_sector = dict(sorted(em_per_sector.items(), key=lambda x: x[1])) # , reverse=True


fig = px.bar(y=em_per_sector.keys(), x=em_per_sector.values(), orientation='h')
fig.update_layout(title='2023 GHG Emissions Per Sector (MtCO2e)',
                  title_x=0.5,
                  xaxis_title="Gas Emissions",
                  yaxis_title="")
fig.show()
```
